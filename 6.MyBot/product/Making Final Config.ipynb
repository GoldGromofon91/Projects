```python
# merge intents by names
final_intents = {}

for config in configs:
    for intent, data in config['intents'].items():
        intent = intent.lower()
        if intent not in final_intents:
            final_intents[intent] = {'examples': [], 'responses': []}
        final_intents[intent]['examples'] += data['examples']
        final_intents[intent]['responses'] += data['responses']

print('Different intent names:', len(final_intents))
```

```python
# remove repeats and empty strings
def clear(intents):
    for intent, data in intents.items():
        data['examples'] = list(set(data['examples']))
        data['responses'] = list(set(data['responses']))
        if '' in data['examples']:
            data['examples'].remove('')
        if '' in data['responses']:
            data['responses'].remove('')

clear(final_intents)
```

```python
# find intents with different names but same examples
intersecting_pairs = set()

for intent1, data1 in final_intents.items():
    for intent2, data2 in final_intents.items():
        if (intent2, intent1) not in intersecting_pairs and intent1 != intent2 \
                and set(data1['examples']) & set(data2['examples']):
            print(intent1, intent2, set(data1['examples']) & set(data2['examples']))
            intersecting_pairs.add((intent1, intent2))
```

```python
# make fixes

def remove(intent, phrases):
    for phrase in phrases:
        final_intents[intent]['examples'].remove(phrase)

def merge(a, b):
    final_intents[a]['examples'] = list(set(final_intents[a]['examples'] + final_intents[b]['examples']))
    final_intents[a]['responses'] = list(set(final_intents[a]['responses'] + final_intents[b]['responses']))
    final_intents.pop(b)

# final_intents.pop('')

remove('music', ['Шалом'])
remove('thanks', ['До свидания'])
remove('whattolisten', ['У тебя есть любимая песня?', 'Какая твоя любимая группа?'])
remove('who_is_author', ['слова песни'])
remove('hip_hop', ['Какую музыку ты любишь?'])

# merge('', '')
```
